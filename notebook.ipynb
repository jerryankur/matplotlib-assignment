# Level 1: Basic Plotting

###  Plot a simple line plot for the function f(x) = x2 using matplotlib

```python
from matplotlib.pyplot import figure, plot, xlabel, ylabel, title, legend, grid, show, hist

from numpy import linspace

# Generate x values
x = linspace(-10, 10, 100)
# Compute f(x) = x^2
y = x ** 2

# Plot
figure(figsize=(8, 6))
plot(x, y, label=r'$f(x) = x^2$', color='blue')
xlabel("x")
ylabel("f(x)")
title("Line Plot of f(x) = x^2")
legend()
grid(True)
show()
```

### Plot a bar graph showing the population of 5 countries

```python
from matplotlib.pyplot import bar

# Data
countries = ['China', 'India', 'United States', 'Indonesia', 'Pakistan']
populations = [1425000000, 1420000000, 336997000, 281604000, 245000000]

# Plot
figure(figsize=(10, 6))
bar(countries, populations, color=['red', 'green', 'blue', 'orange', 'purple'])
xlabel('Country')
ylabel('Population')
title('Population of Selected Countries (2024)')
show()
```

### Create a horizontal bar graph comparing the average temperatures of 4 cities

```python
from matplotlib.pyplot import barh

cities = ['New York', 'Tokyo', 'Sydney', 'Cairo']
average_temperatures = [13.0, 16.0, 18.0, 22.0]  # Average temperatures in Celsius

# Plotting horizontal bar chart
figure(figsize=(10, 6))
barh(cities, average_temperatures, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
xlabel('Average Temperature (C)')
ylabel('City')
title('Average Temperatures of Selected Cities')
grid(axis='x', linestyle='--', alpha=0.7)
show()
```

### Plot a histogram of 1000 random values drawn from a normal distribution using numpy.

```python
from numpy.random import normal

# Generating random values from a normal distribution
data = normal(loc=0, scale=1, size=1000)

# Plotting the histogram
figure(figsize=(8, 6))
hist(data, bins=30, color='purple', edgecolor='black', alpha=0.7)
xlabel('Value')
ylabel('Frequency')
title('Histogram of 1000 Random Values from a Normal Distribution')
grid(axis='y', linestyle='--', alpha=0.7)
show()
```

### Create a scatter plot for two sets of random data points.

```python
from matplotlib.pyplot import scatter
from numpy.random import rand

# Generating two sets of random data points
x = rand(100) * 100  # Random data points for x-axis
y = rand(100) * 100  # Random data points for y-axis

# Plotting scatter plot
figure(figsize=(8, 6))
scatter(x, y, color='teal', edgecolor='black', alpha=0.75)
xlabel('X Values')
ylabel('Y Values')
title('Scatter Plot of Two Sets of Random Data Points')
grid(True)
show()
```

```python
from matplotlib.pyplot import pie

# Data for the pie chart
fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes']
percentages = [30, 20, 25, 25]

# Plotting the pie chart
figure(figsize=(8, 8))
pie(percentages, labels=fruits, autopct='%1.1f%%', startangle=140, colors=['red', 'yellow', 'orange', 'purple'])
title('Percentage of Various Fruits in a Basket')
show()
```

### Plot the sine and cosine functions on the same graph

```python
from numpy import sin, cos, pi

# Generate x values from 0 to 2
x = linspace(0, 2 * pi, 100)
# Calculate sine and cosine values
y_sin = sin(x)
y_cos = cos(x)

# Plotting both sine and cosine functions
figure(figsize=(10, 6))
plot(x, y_sin, label='sin(x)', color='blue')
plot(x, y_cos, label='cos(x)', color='red')
xlabel("x")
ylabel("y")
title("Sine and Cosine Functions")
legend()
grid(True)
show()
```

### Create a stem plot for the function f(x) = sin(x) from 0 to 2

```python
from matplotlib.pyplot import stem

# Generate x values from 0 to 2
x = linspace(0, 2 * pi, 50)
# Calculate f(x) = sin(x)
y = sin(x)

# Creating a stem plot
figure(figsize=(10, 6))
stem(x, y, linefmt='b-', markerfmt='bo', basefmt="r-")
xlabel("x")
ylabel("f(x)")
title("Stem Plot of f(x) = sin(x)")
grid(True)
show()
```

### Plot a basic box plot using a dataset generated from numpy.random.randn()

```python
from matplotlib.pyplot import boxplot
from numpy.random import randn

# Generate random data using numpy's normal distribution
data = randn(100)

# Creating a box plot
figure(figsize=(8, 6))
boxplot(data, vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
ylabel("Values")
title("Box Plot of Random Data from Normal Distribution")
grid(True, axis='y', linestyle='--', alpha=0.7)
show()
```

### Create a stacked bar plot comparing sales of 3 products across 4 quarters

```python
from numpy import array

# Sample data for sales of 3 products across 4 quarters
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
product_A_sales = [150, 200, 250, 300]
product_B_sales = [180, 220, 260, 310]
product_C_sales = [130, 170, 200, 240]

# Plotting stacked bar plot
figure(figsize=(10, 6))
bar(quarters, product_A_sales, label='Product A', color='skyblue')
bar(quarters, product_B_sales, bottom=product_A_sales, label='Product B', color='salmon')
bar(quarters, product_C_sales, bottom=array(product_A_sales) + array(product_B_sales), label='Product C',
    color='lightgreen')

xlabel('Quarters')
ylabel('Sales')
title('Stacked Bar Plot of Product Sales Across Quarters')
legend()
show()
```

### Create a filled area plot for two overlapping sine and cosine curves

```python
from matplotlib.pyplot import fill_between

# Generate x values from 0 to 2
x = linspace(0, 2 * pi, 100)
# Calculate sine and cosine values
y_sin = sin(x)
y_cos = cos(x)

# Plotting filled area plot
figure(figsize=(10, 6))
fill_between(x, y_sin, color='blue', alpha=0.5, label='sin(x)')
fill_between(x, y_cos, color='red', alpha=0.5, label='cos(x)')
xlabel("x")
ylabel("y")
title("Filled Area Plot of Sine and Cosine Curves")
legend()
grid(True)
show()
```

### Use matplotlib to create a step plot for the function f(x) = sin(x)

```python
from matplotlib.pyplot import step

# Generate x values from 0 to 2
x = linspace(0, 2 * pi, 100)
# Calculate f(x) = sin(x)
y = sin(x)

# Creating a step plot
figure(figsize=(10, 6))
step(x, y, where='mid', color='blue', label=r'$f(x) = \sin(x)$')
xlabel("x")
ylabel("f(x)")
title("Step Plot of f(x) = sin(x)")
legend()
grid(True)
show()
```

### Create a plot with multiple subplots showing line plots of sine and cosine in two different subplots.

```python
from matplotlib.pyplot import tight_layout, subplots

# Generate x values from 0 to 2
x = linspace(0, 2 * pi, 100)
# Calculate sine and cosine values
y_sin = sin(x)
y_cos = cos(x)

# Creating subplots
fig, (ax1, ax2) = subplots(2, 1, figsize=(10, 8))

# Sine plot in the first subplot
ax1.plot(x, y_sin, color='blue', label='sin(x)')
ax1.set_title("Sine Function")
ax1.set_xlabel("x")
ax1.set_ylabel("sin(x)")
ax1.legend()
ax1.grid(True)

# Cosine plot in the second subplot
ax2.plot(x, y_cos, color='red', label='cos(x)')
ax2.set_title("Cosine Function")
ax2.set_xlabel("x")
ax2.set_ylabel("cos(x)")
ax2.legend()
ax2.grid(True)

# Adjust layout for clarity
tight_layout()
show()
```

### Plot two line graphs on the same figure, each with its own y-axis (using twinx()).


```python
fig, ax1 = subplots(figsize=(8, 6))
x = linspace(0, 10, 100)
y1 = sin(x)
y2 = cos(x)
ax1.plot(x, y1, color='blue', label="sin(x)")
ax1.set_ylabel("sin(x)", color='blue')
ax2 = ax1.twinx()
ax2.plot(x, y2, color='red', label="cos(x)")
ax2.set_ylabel("cos(x)", color='red')
title("Dual Y-Axis Plot")
show()
```

### Plot a 3D surface plot using data generated from a meshgrid of x and y values.


```python
from numpy import meshgrid

from numpy import sqrt

x = linspace(-5, 5, 50)
y = linspace(-5, 5, 50)
X, Y = meshgrid(x, y)
Z = sin(sqrt(X**2 + Y**2))
fig = figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
fig.colorbar(surf)
title("3D Surface Plot")
show()
```
